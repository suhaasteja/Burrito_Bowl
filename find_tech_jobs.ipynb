```python
# ADVANCED Agent with Structured Output

import os
import json
import nest_asyncio
from typing import List


from langchain_openai import ChatOpenAI
from langchain.agents import AgentExecutor
from langchain.agents.format_scratchpad import format_to_openai_function_messages
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.agents import AgentActionMessageLog, AgentFinish
from pydantic import BaseModel, Field

# Playwright toolkit
from langchain_community.agent_toolkits import PlayWrightBrowserToolkit
from langchain_community.tools.playwright.utils import create_async_playwright_browser

# -------------------------------------------------------------------
# Setup
# -------------------------------------------------------------------
nest_asyncio.apply()

# -------------------------------------------------------------------
# 1) Define the structured response schema (Pydantic)
#    (Docs pattern: use a Pydantic model & bind as a function)
# -------------------------------------------------------------------
class Job(BaseModel):
    entity: str = Field(description="Company name", default="")
    title: str = Field(description="Direct job title", default="")
    url: str = Field(description="Direct job URL", default="")
    date_posted: str = Field(description="Posting date as displayed on the page", default="")
    location: str = Field(description="Job location text", default="")

class JobListResponse(BaseModel):
    """Final structured response containing extracted job postings"""
    jobs: List[Job] = Field(
        description="List of extracted jobs with fields: entity, url, date_posted, location"
    )

# -------------------------------------------------------------------
# 2) Build Playwright tools
# -------------------------------------------------------------------
async_browser = create_async_playwright_browser(headless=True, )
toolkit = PlayWrightBrowserToolkit.from_browser(async_browser=async_browser)
tools = toolkit.get_tools()  # e.g., navigate_browser, get_elements, extract_text, etc.

# -------------------------------------------------------------------
# 3) LLM + bind tools and response schema as functions
# -------------------------------------------------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0, api_key=OPENAI_API_KEY)

# IMPORTANT: bind BOTH the tools and the structured response schema
llm_with_functions = llm.bind_functions([*tools, JobListResponse])

# -------------------------------------------------------------------
# 4) Custom parser (from docs):
#    - If the model calls our JobListResponse "function", finish with those args
#    - If it calls a tool, return AgentActionMessageLog to run that tool
#    - If no function called, treat message content as final text (fallback)
# -------------------------------------------------------------------
def parse(output):
    # If no function was invoked, return text content (fallback)
    if "function_call" not in output.additional_kwargs:
        return AgentFinish(return_values={"output": output.content}, log=output.content)

    function_call = output.additional_kwargs["function_call"]
    name = function_call["name"]
    inputs = json.loads(function_call["arguments"] or "{}")

    # If the JobListResponse function is invoked, we are done: return structured output
    if name == "JobListResponse":
        return AgentFinish(return_values=inputs, log=str(function_call))

    # Otherwise, it's a tool call: run the tool
    return AgentActionMessageLog(
        tool=name,
        tool_input=inputs,
        log="",
        message_log=[output],
    )
    
import asyncio
import traceback

async def safe_invoke(agent_executor, inp, company, keyword, url):
    try:
        out = await agent_executor.ainvoke({"input": inp}, return_only_outputs=True)
        jobs = out.get("jobs", [])
    except Exception as e:
        # Log or print the error for debugging
        print(f"[WARN] Error {url} {company} {keyword}")
        traceback.print_exc()

        # Return a default empty JobListResponse
        jobs = [{
            "entity": company,
            "title": "",
            "url": url,
            "date_posted": "",
            "location": "",
            "company": company,
            "search_keyword": keyword,
        }]
    return jobs


# -------------------------------------------------------------------
# 5) Prompt & Agent graph (docs pattern)
# -------------------------------------------------------------------
prompt = ChatPromptTemplate.from_messages(
    [
        ("system",
         "You are a careful browsing agent. Use the provided browser tools to navigate and extract data. "
         "When you are done, return the final result by calling the JobListResponse function with the JSON fields requested."
         "If you come across an error such as TimeoutError, return empty vals for keys in JobListResponse"),
        ("user", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = (
    {
        "input": lambda x: x["input"],
        # Format intermediate steps into OpenAI function messages
        "agent_scratchpad": lambda x: format_to_openai_function_messages(
            x.get("intermediate_steps", [])
        ),
    }
    | prompt
    | llm_with_functions
    | parse
)

agent_executor = AgentExecutor(
    tools=tools,
    agent=agent,
    verbose=False,
)
```

```python
import pandas as pd
import json
from urllib.parse import quote_plus
from tqdm import tqdm

# CSV: columns company,url (url has {search_keyword} placeholder)
df = pd.read_csv("fortune_500_careers_final.csv")

# Keywords to search
search_keywords = ["ai engineer", "data scientist"]

all_jobs = []

for _, row in tqdm(df.iterrows(), total=df.shape[0], desc="searching.."):
    company = str(row.get("Company Name", "")).strip()
    url_template = str(row.get("updated_url", "")).strip()
    if not url_template:
        continue

    for keyword in search_keywords:
        # URL-encode the keyword for query params
        encoded_kw = quote_plus(keyword)
        url = url_template.format(search_keyword=encoded_kw)

        inp = (
            f"Go to {url} . "
            "Use the browser tools to extract multiple jobs with fields: "
            "entity, url, date_posted, location. "
            "When finished, call JobListResponse with the results."
        )

        # out = await agent_executor.safe({"input": inp}, return_only_outputs=True)
        jobs = await safe_invoke(agent_executor, inp, company, keyword, url)
        for j in jobs:
            j["company"] = company
            j["search_keyword"] = keyword
        all_jobs.extend(jobs)
        # print(all_jobs)

# dedupe by URL
seen = set()
deduped = []
for j in all_jobs:
    u = (j.get("url") or "").strip()
    if u and u not in seen:
        seen.add(u)
        deduped.append(j)

# print(json.dumps(deduped, ensure_ascii=False, separators=(",", ":")))
```
